```python
#Write a function that determines the number of digits in an integer

def num_digits(p: int) -> int:
    num_digits = 0
    while (p != 0):
      p //= 10
      num_digits += 1
    return num_digits
num = 3242392529
print(f"the amount of digits are: ",num_digits(num) )
```

```python
#A three digit number is called Armstrong, if it is equual to the sum of the cubes of its digits:
#Example:  156=13+53+63=1+125+216=342≠156  is NOT an Armstrong number
#But  153=13+53+33=1+125+27=153  is an Armstrong number.
#Write a function to check if a given three digit number is Armstrong

def is_armstrong(n: int) -> bool:
    n_str = str(n)
    d = len(n_str)
    total = sum(int(digit)**d for digit in n_str)
    return total == n

n = 153
if (is_armstrong(n)):
  print(f"{n} is an armstrong number ")
else:
  print(f"{n} is not an armstrong")
```

```python
#prime number stuffs

def biggest_prime_of_eight_digit_number(num: int) -> int:
    def is_prime(num: int) -> bool:
        if num < 2:
            return False
        for i in range(2, num):
            if num % i == 0:
                return False
        return True
    num = list(str(num))
    num.sort(reverse=True)
    N = len(num)
    res = []
    for i in range(N):
        if is_prime(int(num[i])):
            res.append(num[i])
    return int("".join(res))

num = 2
if(biggest_prime_of_eight_digit_number (num)):
 print(f"the number {num} is prime")
else:
 print(f"the {num} is not prime")
```

```python
def is_valid_triangle(side1, side2, side3):
  """
  Checks if the given three sides form a valid triangle.

  Args:
    side1: The length of the first side.
    side2: The length of the second side.
    side3: The length of the third side.

  Returns:
    True if the given sides form a valid triangle, False otherwise.
  """

  # Check if any side is less than or equal to 0.
  if side1 <= 0 or side2 <= 0 or side3 <= 0:
    return False

  # Check if the sum of any two sides is less than or equal to the third side.
  if side1 + side2 <= side3 or side1 + side3 <= side2 or side2 + side3 <= side1:
    return False

  # If all the above conditions are not met, then the given sides form a valid triangle.
  return True

#this actually runs the code
side1 = 5
side2 = 7
side3 = 10
if (is_valid_triangle(side1,side2,side3)):
  print(side1,"",side2,"",side3, "is True")
else:
  print("not a valid triangle")
```

```python
def kaprekar_routine(num):
  """
  Performs the Kaprekar routine on a four-digit number.

  Args:
    num: A four-digit number with at least two different digits.

  Returns:
    The final number after the routine stops.
  """

  while True:
    # Convert the number to a string to perform digit manipulation.
    num_str = str(num)

    # Arrange the digits in descending and ascending order.
    descending_num = int(''.join(sorted(num_str, reverse=True)))
    ascending_num = int(''.join(sorted(num_str)))

    # Subtract the smaller number from the bigger number.
    num = descending_num - ascending_num

    # Check if the routine has stopped.
    if descending_num == ascending_num:
      return num

num = 1945
if kaprekar_routine(num):
  print("the final number after the Kaprekar routine is",num)
```

```python
def generate_sequence(start_num):
  """Generates the sequence ending in 4, 2, 1 for the given starting number.

  Args:
    start_num: The starting number of the sequence.

  Returns:
    A list containing the sequence.
  """

  sequence = [start_num]

  while sequence[-1] != 1:
    if sequence[-1] % 2 == 0:
      next_num = sequence[-1] // 2
    else:
      next_num = 3 * sequence[-1] + 1
    sequence.append(next_num)

  return sequence

start_num = 7
if generate_sequence(start_num):
  print(generate_sequence(start_num))
```

```python
#Print integers 1 to N, but print “Fizz” if an integer is divisible by 3,
# “Buzz” if an integer is divisible by 5,
#“FizzBuzz” if an integer is divisible by both 3 and 5.

def fizz_buzz(n: int) -> list[str]:
    result = []
    for i in range(1, n + 1):
        if i % 3 == 0 and i % 5 == 0:
            result.append("FizzBuzz")
        elif i % 3 == 0:
            result.append("Fizz")
        elif i % 5 == 0:
            result.append("Buzz")
        else:
            result.append(str(i))
    return result

n = 15
output = fizz_buzz(n)
print(output)
```

```python
#If the lowercase version of the character is a vowel (i.e., ‘a’, ‘e’, ‘i’, ‘o’, or ‘u’), we convert it to uppercase.
#Otherwise, we convert it to lowercase.
#The resulting string is stored in the result variable and returned.
#set vowels to uppercase and consonants to upper

def vowel_to_upper_cons_to_lower(string: str)-> str:
    result = ""
    for char in string:
      if char.lower() in "aeiou":
        result += char.upper()
      else:
        result += char.lower()
    return result

input_string = "hello python world im tired"
output_string = vowel_to_upper_cons_to_lower(input_string)
print(output_string)

```

```python
#Program that separates evens and odds in a list

def split_Even_Odd_Sublist(num_array):
  ev_li = [ele for ele in num_array if ele % 2 == 0]
  od_li = [ele for ele in num_array if ele % 2 != 0]
  return ev_li, od_li

numbers = [2,5,10,13,17,20]
even_list , odd_list = split_Even_Odd_Sublist(numbers)
print("Evens: ", even_list)
print("Odds: ",odd_list)
```
